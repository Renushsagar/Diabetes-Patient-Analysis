```python
# Assuming 'Outcome' is the target variable
sns.countplot(x='Outcome', data=df)

# Get the counts for each category
counts = df['Outcome'].value_counts()

# Add labels to the bars
for i, count in enumerate(counts):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.title('Distribution of Outcome')
plt.show()
```

```python
# Assuming 'Glucose' is the variable you're plotting
plt.figure(figsize=(8, 6))
sns.histplot(df['Glucose'], kde=True)

# Add labels to the bars
counts, bins, patches = plt.hist(df['Glucose'], bins=30, color='skyblue', edgecolor='black')

for count, bin, patch in zip(counts, bins, patches):
    plt.text(bin + 0.5, count + 1, f'{int(count)}', ha='center')

plt.title('Distribution of Glucose')
plt.show()
```
